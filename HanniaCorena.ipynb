```python
inventario = []

inventario = [
{'nombre': 'Camisa', 'precio': 20, 'cantidad': 40},
{'nombre': 'Pantalón', 'precio': 30, 'cantidad': 30},
{'nombre': 'Zapatos', 'precio': 50, 'cantidad': 25},
{'nombre': 'Chaqueta', 'precio': 70, 'cantidad': 15},
{'nombre': 'Falda', 'precio': 25, 'cantidad': 20},
{'nombre': 'Gorra', 'precio': 10, 'cantidad': 50},
{'nombre': 'Bufanda', 'precio': 15, 'cantidad': 35},
{'nombre': 'Calcetines', 'precio': 5, 'cantidad': 100},
{'nombre': 'Cinturón', 'precio': 12, 'cantidad': 40},
{'nombre': 'Bolso', 'precio': 45, 'cantidad': 18}
]

def agregar_producto(nombre, precio, cantidad):
    for producto in inventario:
        if producto['nombre'].lower() == nombre.lower():
            producto['cantidad'] += cantidad
            print(f"Cantidad actualizada de {nombre}. Nueva cantidad: {producto['cantidad']}")
            return
        nuevo_producto = {
        'nombre': nombre,
        'precio': precio,
        'cantidad': cantidad
    }
    inventario.append(nuevo_producto)
    print(f"Producto agregado: {nombre}")

agregar_producto("Botines", 20, 2)  
agregar_producto("Gorra", 10, 50)   

print("\nInventario actual:")
print(inventario)
```

```python
def ver_inventario():
    print("\nInventario de productos:")
    for producto in inventario:
        print(f"Nombre: {producto['nombre']}, Precio: ${producto['precio']}, Cantidad: {producto['cantidad']}")
ver_inventario()
```

```python
def eliminar_producto(nombre):
    for i, producto in enumerate(inventario):
        if producto["nombre"].lower() == nombre.lower():
            eliminado = inventario.pop(i)
            print(f"Producto eliminado: {eliminado['nombre']}")
            return
    print("El producto no está en el inventario.")
eliminar_producto("Cinturón")
```
